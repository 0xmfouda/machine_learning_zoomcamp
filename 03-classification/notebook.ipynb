#### Transpose the dataframe to take a look at the columns

```python
df.head().T
```

#### Normalize Names

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
# For columns with string values, lowercase the values and replace spaces with underscores.
categorical_names = list(df.dtypes[df.dtypes == 'object'].index)

for c in categorical_names:
  df[c] = df[c].str.lower().str.replace(" ", "_")
```

#### Check & Fix Data Types

```python
df.dtypes
```

```python
tc = pd.to_numeric(df.totalcharges, errors='coerce')
```

```python
tc.isnull().sum()
```

```python
df[tc.isnull()][['customerid', 'totalcharges']]
```

```python
df.totalcharges = pd.to_numeric(df.totalcharges, errors='coerce')
df.totalcharges = df.totalcharges.fillna(0)
```

#### Convert `Churn` from "Yes/No" to `1/0`

```python
df.churn.head()
```
